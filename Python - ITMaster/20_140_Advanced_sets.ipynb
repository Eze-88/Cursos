```python
# Creamos un set y le asignamos items
s = set()
s = {1,2,3}

# Creamos otro set
sc = s.copy()
s.add(4)
print(sc)
print(s)
```

```python
#Podemos ver las diferencias entre sets
print(s.difference(sc))

# Devolver todos los elementos restantes luego de que se eliminan los que hay en común
s1 = {1,2,3}
s2 = {1,4,5}
s1.difference_update(s2)
print(s1)
```

```python
# Podemos descartar elementos si existen en el set
s = {1,2,3,4}
s.discard(2)
print(s)
s.discard(6)
print(s)
```

```python
# Se consiguen los elementos en comun entre dos sets
s1 = {1,2,3}
s2 = {1,2,4}

print(s1.intersection(s2))

#Se puede actualizar s1 para que sea la interseccion
s1.intersection_update(s2)
print(s1)
```

```python
# Devuelve verdadero/falso si son independientes
s1 = {1,2,3}
s2 = {1,2,4}
s3 = {5}

#Va a dar falso porque hay elementos en común
print(s1.isdisjoint(s2))
```

```python
# Es un set un sub-set de otro?
s1 = {1,2}
s2 = {1,2,4}
print(s1.issubset(s2))
#lo mismo pero al revés, super-set
print(s2.issuperset(s1))
```
