```python
import csv

#Abrir el archivo
data = open('/content/example.csv')

#Leer el csv
csv_data = csv.reader(data)

#fomarteamos el archivo a un objeto de python
data_lines = list(csv_data)
```

```python
#Vemos que información trajimos
print(data_lines[0])
print(len(data_lines))
```

```python
#Para ver los primeros 5 items
for line in data_lines[:5]:
  print(line)
```

```python
#Podemos hacer una lista con los correos
all_emails = []

for line in data_lines[1:5]:
  all_emails.append(line[3])

print(all_emails)
```

```python
#Hacemos una lista de nombre y apellidos
full_names = []

for line in data_lines[1:5]:
  full_names.append(line[1] + ' ' + line[2])

print(full_names)
```

```python
#Guardamos la lista de nombres completos

file_to_output = open('to_save_file.csv',mode='w',newline='')
```

```python
#el delimitador es un sepador arbitrario
csv_writer = csv.writer(file_to_output,delimiter=',')
```

```python
#podemos escribir de a una columna
csv_writer.writerow(['a','b','c'])
```

```python
#se depuede escribir de a varias columnas
csv_writer.writerows([['1','2','3'],['4','5','6']])
```

```python
#cerramos el arhivo
file_to_output.close()
```

```python
import PyPDF2

#abrimos el arvhivo en modo lectura binaria
f = open('Working_Business_Proposal.pdf','rb')
```

```python
#cargamos el archivo en una variable para probar
pdf_reader = PyPDF2.PdfFileReader(f)

#podmemos ver cuántas paginas tiene
print(pdf_reader.numPages)

#podemos cargar de a una pagina, emepzando por el 0, y ver el contenido
page_one = pdf_reader.getPage(0)
page_one_text = page_one.extractText()
#hace falta que el texto exista y sea legible
print(page_one_text)
f.close()
```

```python
#Agregando paginas al pdf
#Podemos tomar la primer pagina de unos PDFs dados y agregarlos a uno nuevo para separarlos

#abrimos el arvhivo en modo lectura binaria
f = open('Working_Business_Proposal.pdf','rb')
pdf_reader = PyPDF2.PdfFileReader(f)
```

```python
page_one = pdf_reader.getPage(0)
```

```python
pdf_writer = PyPDF2.PdfFileWriter()
pdf_writer.addPage(page_one)
```

```python
#Probamos que se haya agregado bien
#este comando crea (y sobreescribe si ya existe) un PDF
pdf_output = open('Archivo_PDF_Nuevo.pdf','wb')
pdf_writer.write(pdf_output)
f.close()
pdf_output.close()
```
