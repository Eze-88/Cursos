```python
# Generamos un diccionario
d = {'k1':1,'k2':2}
print(d)
```

```python
# Podemos generar automaticamente un diccionario
print({x:x**2 for x in range(10)})
print({k:v**2 for k,v in zip(['a','b'],range(2))})
```

```python
# Los diccionarios se pueden iterar
for k in d.items():
  print(k)
```
