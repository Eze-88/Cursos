```python
# Creamos unas listas
l = [1,2,3,4]
print(l)
```

```python
# Agregamos una lista a la otra, pero lo agrega como un nuevo elemento
x = [1,2,3]
x.append([4,5])
print(x)
#Esto etiende una lista
x.extend([4,5])
print(x)
```

```python
# Insertar un elemento en un lugar especifico
print(l)
l.insert(2,'insertado')
print(l)
```

```python
# Remover la primera ocurrencia de un valor
l = [1,2,3,4,1]
print(l)
l.remove(1)
print(l)
# Damos vuelta la lsta
l.reverse()
print(l)
#ordenamos la lista
l.sort()
print(l)
```
