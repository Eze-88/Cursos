```python
# Cambiar mayusculas/minusculas
s = 'hello world'
print(s.capitalize())
print(s.upper())
print(s.lower())
```

```python
#Contar caracteres
print(s.count('o'))
#Encontrar la primera aparicion de un caracter
print(s.find('o'))
```

```python
# Podemos poner una cadena en el medio de otra cadena
print(s.center(20,'z'))
```

```python
#Expandir caracteres especiales sin tener que imprimirlo
'hello\thi'.expandtabs()
```

```python
#Verificacion de que todos los caracteres sean alfanumericos
s = 'hello'
print(s.isalnum())
#Verificacion de que todos los caracteres sean alfanumericos
print(s.isalpha())
```

```python
#Son todos los caracteres minusculas/mayuscula?
print(s.islower())
print(s.isupper())
#Son todos los caracteres espacios en blanco?
print(s.isspace())
#Si termina con una letra especifica
print(s.endswith('o'))
```

```python
#Dividir un string en un caracter
print(s.split('e'))

s = 'hihhiihiiihhii'
print(s.split('i'))

# Se puede hacer ante la primera aparicion del caracter
print(s.partition('i'))
```
